```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
import gradio as gr
```

```python
load_dotenv()

openai = OpenAI();
```

```python
system_message = "You are an helpful Assistant."
def message_gpt(prompt):
    messages = [{"role":"system","content":system_message},{"role":"user","content":prompt}]
    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
print(message_gpt("hey !"))
```

```python
def shout(text):
    print("Hey i am shouting at you")
    return text.upper();
```

```python
message_input = gr.Textbox(label="Your message:", info="Enter a message to be shouted", lines=7)
message_output = gr.Textbox(label="Response:", lines=8)

view = gr.Interface(
    fn=message_gpt,
    title="Shout", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=["hello", "howdy"], 
    flagging_mode="never"
    )
view.launch()
```

```python
system_message = "You are helful assistant that always reply in mark down"

message_input = gr.Textbox(label="Your message:", info="Enter a message to be shouted", lines=7)
message_output = gr.Markdown(label="Response")

view = gr.Interface(
    fn=message_gpt,
    title="Shout", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=["hello", "howdy"], 
    flagging_mode="never"
    )
view.launch()
```

```python
def stream_gpt(prompt):
    messages = [{"role":"system","content":system_message},{"role":"user","content":prompt}]
    view = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in view:
        result += chunk.choices[0].delta.content or ""
        yield result


gr.Interface(fn=stream_gpt,title="GPT",inputs=message_input,outputs=message_output,flagging_mode="never").launch()
```
